```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('health.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# SQL query to create the 'patient_data' table
create_table_query = '''
CREATE TABLE IF NOT EXISTS patient_data (
    patient_id INTEGER PRIMARY KEY,
    patient_name TEXT,
    age INTEGER,
    gender TEXT,
    diagnosis TEXT
);
'''

# Execute the CREATE TABLE query
cursor.execute(create_table_query)

# SQL query to insert sample data into the table
insert_data_query = '''
INSERT INTO patient_data (patient_name, age, gender, diagnosis)
VALUES
    ('Jenny Pur', 35, 'Female', 'Asthma'),
    ('Janet Hicks', 28, 'Female', 'Diabetes');
'''

# Execute the INSERT INTO query
cursor.execute(insert_data_query)

# Commit the changes and close the database connection
conn.commit()
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('health.db')

# List tables
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables in the database:")
for table in tables:
    print(table[0])

# Check table structure
cursor.execute("PRAGMA table_info(patient_data);")
table_info = cursor.fetchall()
print("Structure of 'patient_data' table:")
for col in table_info:
    print(col)

# Retrieve data from the table
cursor.execute("SELECT * FROM patient_data;")
data = cursor.fetchall()
print("Data in 'patient_data' table:")
for row in data:
    print(row)

# Close the connection
conn.close()
```
